### get user list

챌린저, 그랜드마스터, 마스터 티어의 유저 목록

```python
user_url = 'https://kr.api.riotgames.com/lol/league-exp/v4/entries/RANKED_SOLO_5x5/{}/I?page={}&api_key={}'
```

```python
tiers = ['CHALLENGER','GRANDMASTER','MASTER']
```

```python
user_tier = []
for tier in tiers: 
    for page_num in tqdm(range(1,20)):
        # url formatting
        url_now = user_url.format(tier,page_num,my_api_key)
        
        # request
        reqq =  requests.get(url_now).text
        
        if reqq == '[]':
            pass
        else:
            df = json.loads(reqq)
            user_tier.append(pd.DataFrame(df))
            
    time.sleep(30)
```

```python
user_df = pd.concat(user_tier)
user_df = user_df.reset_index(drop=True)
user_df.head(3)
```

```python
user_df.groupby('tier',as_index=False).size()
```

#### eda

```python
user_df['matches'] = user_df['wins'] + user_df['losses']
user_df['win_rt']  = user_df['wins'] / user_df['matches']
```

```python
plt.figure(figsize=(15,5))
plt.hist(user_df['win_rt']);
```

```python
plt.figure(figsize=(15,6))
plt.scatter(x=user_df['matches'],y=user_df['win_rt'],alpha=0.1)
plt.axhline(user_df['win_rt'].mean(),color='red',alpha=0.7)
plt.axvline(user_df['matches'].mean(),color='red',alpha=0.7)
plt.title('플레이어별 승률');
plt.xlabel('게임 판수');
plt.ylabel('승률');
plt.text(1000,0.65,'평균승률 : %.2f%% \n평균판수 : %.0f판' % (user_df['win_rt'].mean() * 100,
                                                           user_df['matches'].mean()));
```

```python
fig,ax = plt.subplots(3,1,figsize=(18,10),
                     sharex=True,sharey=True)
text_position = [4400,4300,4600]
for numm in [0,1,2]:
    tier = tiers[numm]
    dff = user_df[user_df['tier']==tier]
    ax[numm].scatter(x=dff['matches'],y=dff['win_rt'],alpha=0.1)
    ax[numm].axhline(dff['win_rt'].mean(),color='red',alpha=0.7)
    ax[numm].axvline(dff['matches'].mean(),color='red',alpha=0.7)
    ax[numm].set_xticks(range(0,5500,500))
    ax[numm].text(text_position[numm],0.95,tier,fontsize=15)
    ax[numm].text(1250,0.65,'평균승률 : %.2f%% \n평균판수 : %.0f판' % (
        dff['win_rt'].mean() * 100,dff['matches'].mean()),fontsize=12);
```

#### puid

```python
puid_url = 'https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/{}?api_key={}'
```

```python
puid_list = []
i = 0
iter_time = datetime.now() + timedelta(minutes=2)
for roww in tqdm(user_df.iterrows()):
    
    summoner = roww[1]['summonerName']
    urll = puid_url.format(summoner,my_api_key)
    
    if i == 99:
        time.sleep(120)
    
        reqq = requests.get(urll)
        dff = pd.DataFrame(json.loads(reqq.text),index=[0])
        puid_list.append(dff)
        
        i = 0
        
    else:
        reqq = requests.get(urll)
        dff = pd.DataFrame(json.loads(reqq.text),index=[0])
        puid_list.append(dff)
        
        i += 1
```

```python
puid_df = pd.concat(puid_list)
puid_df = puid_df.reset_index(drop=True)
puid_df.head(3)
```

#### bind user & puid

```python
user_puid_df = pd.concat([user_df,puid_df],axis=1)
user_puid_df.head(5)
```

#### fill na's

```python
puid_again = []
for summoner in user_puid_df.loc[user_puid_df['name'].isna(),'summonerName']:
    urll = puid_url.format(summoner,my_api_key)
    
    reqq = requests.get(urll).text
    dff = pd.DataFrame(json.loads(reqq),index=[0])
    dff['return_text'] = reqq
    puid_again.append(dff)
```

```python
puid_df2 = pd.concat(puid_again)
puid_df2.reset_index(drop=True,inplace=True)
puid_df2 = puid_df2.dropna(subset=['id'],axis=0)
puid_df2.head(5)
```

```python
for roww in puid_df2.iterrows():
    user_puid_df.loc[user_puid_df['summonerName']==roww[1]['name'],'id'] = roww[1]['id']
    user_puid_df.loc[user_puid_df['summonerName']==roww[1]['name'],'accountId'] = roww[1]['accountId']
    user_puid_df.loc[user_puid_df['summonerName']==roww[1]['name'],'puuid'] = roww[1]['puuid']
    user_puid_df.loc[user_puid_df['summonerName']==roww[1]['name'],'name'] = roww[1]['name']
    user_puid_df.loc[user_puid_df['summonerName']==roww[1]['name'],'profileIconId'] = roww[1]['profileIconId']
    user_puid_df.loc[user_puid_df['summonerName']==roww[1]['name'],'revisionDate'] = roww[1]['revisionDate']
    user_puid_df.loc[user_puid_df['summonerName']==roww[1]['name'],'summonerLevel'] = roww[1]['summonerLevel']
```
