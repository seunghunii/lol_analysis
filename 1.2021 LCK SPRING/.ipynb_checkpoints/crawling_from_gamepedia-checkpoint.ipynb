```python
def select_cols(str_contain,df,with_patch=True,sel_type='contains'):
    if sel_type == 'contains':
        to_contain = df.columns.str.contains(str_contain)
    
    elif sel_type == 'start':
        to_contain = df.columns.str.startswith(str_contain)
    
    elif sel_type == 'end':
        to_contain = df.columns.str.endswith(str_contain)
    
    cols_list = df.columns[to_contain].values.tolist()
    
    if with_patch == True:
        cols_list.append('Patch')
        return cols_list
    else:
        return cols_list
```

```python
info_cols = ['Phase', 'Blue', 'Red', 'Score', 'Winner','Winner_team']
position_df = ban_pick_df.loc[:,select_cols('Position',ban_pick_df)+info_cols]
ban_df = ban_pick_df.loc[:,select_cols('Ban',ban_pick_df)+info_cols]
pick_df = ban_pick_df.loc[:,select_cols('Pick',ban_pick_df)+info_cols]
```

## 팀별, 진영별 포지션 선픽

살펴볼점.

블루사이드   
대부분의 팀이 블루 1픽에서 정글을 가장 많이 뽑았음.   
볼만한 점은 담원기아는 1픽으로 정글보다 원딜을 조금 더 많이 뽑았음.   
한화는 블루 1픽에서 다른 포지션을 뽑은적이 아예 없음. 정글 원딜을 항상 먼저 뽑음.(정글메타 + 데프트)   
농심은 의외로 블루 1픽에서 서폿을 가장 많이 뽑았다.(켈린)   
KT는 탑미드를 1픽을 뽑은 적이 없음. 정글을 제일 많이 뽑고, 원딜서폿의 비율이 비슷함.   
   
레드사이드   
레드 1픽이라는건 상대 블루 1픽에서 해당 포지션을 뽑지 않았다는 의미.   
아프리카와 한화는 정글을 가장 많이 뽑았음. 두팀 다 미드를 레드 1픽에서 뽑은 적이 없음.   
DRX, 담원, 농심, T1은 원딜을 가장 많이 뽑았음.   
KT는 서폿이 가장 많았고, 이후 원딜, 미드, 정글 순.   
담원과 KT 둘다 레드 1픽으로 탑을 뽑은 적이 없다.   

총 정리.   
모든 팀에서 확인할 수 있는 정글, 원딜 메타.   
그 외에는 전체적으로 벤픽할 때 우선순위에 두는 포지션이 다른 것을 볼 수 있다.

```python
pikcol = select_cols('_Pick',ban_pick_df,with_patch=False)
np.unique(ban_pick_df[pikcol].values.tolist())[0:10]
```

```python
for idx, row in ban_pick_df[pikcol].iterrows():
     ban_pick_df.loc[idx,pikcol] = row.str.strip()
```

```python
np.unique(ban_pick_df[pikcol].values.tolist())[0:10]
```

```python
blue_ban = []
blue_ban_cols = select_cols('Blue_Ban',ban_pick_df,with_patch=False)
for k in range(0,len(ban_df)):
    tmp_list = []
    tmp_list = list(ban_df.loc[k,blue_ban_cols])
    blue_ban.append(sorted(tmp_list))
```

```python
red_ban = []
red_ban_cols = select_cols('Red_Ban',ban_pick_df,with_patch=False)
for k in range(0,len(ban_df)):
    tmp_list = []
    tmp_list = list(ban_df.loc[k,red_ban_cols])
    red_ban.append(sorted(tmp_list))
```

```python
blue_ban[0:5]
red_ban[0:5]
```

```python
te = TransactionEncoder()

te_blue = te.fit_transform(blue_ban)
blue_ban_df = pd.DataFrame(te_blue,columns=te.columns_)

te_red  = te.fit_transform(red_ban)
red_ban_df  = pd.DataFrame(te_red,columns=te.columns_)
```

```python
apriori(blue_ban_df, min_support=0.2, use_colnames=True)
apriori(red_ban_df, min_support=0.2, use_colnames=True)
```

## ban & picks

### blue

```python
blue_banpick = [k for k in select_cols('Blue_',ban_pick_df,with_patch=False) if not k.startswith('Blue_Position')]
blue_banpick_df = ban_pick_df.loc[:,blue_banpick]
blue_banpick_df.head()
```

```python
bancol = select_cols('_Ban',blue_banpick_df,with_patch=False)
pikcol = select_cols('_Pick',blue_banpick_df,with_patch=False)
blue_list = []
for idx,rows in blue_banpick_df.iterrows():
    tmp_list = ['Ban_'+ str(a) for a in rows[bancol]] + ['Pick_'+ str(b) for b in rows[pikcol]]
    blue_list.append(tmp_list)
```

```python
te = TransactionEncoder()

te_blue_banpick = te.fit_transform(blue_list)
te_blue_banpick_df2 = pd.DataFrame(te_blue_banpick,columns=te.columns_)
```

```python
te_blue_result = apriori(te_blue_banpick_df2,min_support=0.1,use_colnames=True)
te_blue_result['length'] = te_blue_result['itemsets'].apply(lambda x : len(x))
te_blue_result.sort_values('length',ascending=False).head(10)
```

```python
association_rules(te_blue_result, metric="confidence", min_threshold=0.01)
```

### red

```python
red_banpick = [k for k in select_cols('Red_',ban_pick_df,with_patch=False) if not k.startswith('Red_Position')]
red_banpick_df = ban_pick_df.loc[:,red_banpick]
red_banpick_df.head()
```

```python
bancol = select_cols('_Ban',red_banpick_df,with_patch=False)
pikcol = select_cols('_Pick',red_banpick_df,with_patch=False)
red_list = []
for idx,rows in red_banpick_df.iterrows():
    tmp_list = ['Ban_'+ str(a) for a in rows[bancol]] + ['Pick_'+ str(b) for b in rows[pikcol]]
    red_list.append(tmp_list)
```

```python
te = TransactionEncoder()

te_red_banpick = te.fit_transform(red_list)
te_red_banpick_df2 = pd.DataFrame(te_red_banpick,columns=te.columns_)
```

```python
te_red_result = apriori(te_red_banpick_df2, min_support=0.1, use_colnames=True)
te_red_result['length'] = te_red_result['itemsets'].apply(lambda x : len(x))
te_red_result.sort_values('length',ascending=False).head(10)
# 레드팀에서는 카이사를 픽하면 레넥, 세라핀을 많이 벤했고,
# 카이사를 픽하면 신드라도 많이 픽했다.
```

```python
association_rules(te_red_result, metric="confidence", min_threshold=0.01)
```
